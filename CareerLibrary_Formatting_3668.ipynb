```python
import re
import csv

output_file = "split_paths.csv"

# Open the file for writing and initialize CSV writer
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write header row once
    writer.writerow([f'Path {i + 1}' for i in range(4)])  # Assuming up to Path 4

    # Process each row in the DataFrame
    for index, input_content in enumerate(df["Education Path"]):
        # Check if input_content is a string before applying regex
        if isinstance(input_content, str):
            # Split content into paths using regex
            paths = re.split(r'path\d+-', input_content)
            paths = [path.strip() for path in paths if path.strip()]  # Remove empty and whitespace-only entries
        else:
            # Handle non-string values (e.g., NaN) by assigning an empty list
            paths = ['']  # or any other default value

        # Ensure rows align with columns (pad with empty strings if necessary)
        while len(paths) < 4:  # Adjust 4 to the maximum number of paths expected
            paths.append('')

        # Write content to CSV
        writer.writerow(paths)

print(f"Paths successfully split and saved to '{output_file}'")
```
