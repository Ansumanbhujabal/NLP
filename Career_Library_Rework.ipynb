```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import numpy as np

# Load the data
file_path = "your_file.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Initialize the SentenceTransformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Extract columns
new_names = df["NEW"].dropna().tolist()  # Exclude NaN values
career_names = df["Career Name"].tolist()

# Encode the career names and new names into vectors
new_name_embeddings = model.encode(new_names, convert_to_tensor=True)
career_name_embeddings = model.encode(career_names, convert_to_tensor=True)

# Store results
results = []

# Set a similarity threshold
similarity_threshold = 0.7  # Adjust as needed

# For each new name, find matches in the career names
for i, new_name in enumerate(new_names):
    # Check for exact match first
    if new_name in career_names:
        matched_row = df[df["Career Name"] == new_name].iloc[0]
        results.append(matched_row)
    else:
        # Compute cosine similarity
        similarities = util.pytorch_cos_sim(new_name_embeddings[i], career_name_embeddings)

        # Get the index of the most similar career
        top_match_idx = np.argmax(similarities[0])
        top_similarity = similarities[0][top_match_idx].item()

        if top_similarity >= similarity_threshold:
            matched_row = df.iloc[top_match_idx]
            results.append(matched_row)

# Create a DataFrame from the matched rows
matched_df = pd.DataFrame(results)


# Save the updated DataFrame to a new CSV file
output_file = "updated_file.csv"  # Replace with your desired output file name
df.to_csv(output_file, index=False)

print(f"Matching and updating complete! Results saved to {output_file}")
```
